```julia
using BSeries
using RootedTrees
using Plots
```

# Printing the energy-preserving conjugates

```julia
for tree in RootedTreeIterator(7)
    println(tree.level_sequence)
    println("Conjugates:")
    for t in BSeries.equivalent_trees(tree.level_sequence)
        energy_preserving_conjugate = RootedTree(BSeries.rightmost_energy_preserving_tree(t))
        if (-1)^BSeries.num_branches(t) == -1
            print("-")
        else
            print("+")
        end
        if energy_preserving_conjugate == rootedtree(t)
            println("self")
        else
            println(energy_preserving_conjugate.level_sequence)
        end
    end
    println("")
end
```

# Printing each subset only once

## Order 4

```julia
treelist = []
for tree in RootedTreeIterator(4)
    if !(tree in treelist)
        push!(treelist,copy(tree))
        println(tree.level_sequence)
        println("Conjugates:")
        for t in BSeries.equivalent_trees(tree.level_sequence)
            energy_preserving_conjugate = RootedTree(BSeries.rightmost_energy_preserving_tree(t))
            if (-1)^BSeries.num_branches(t) == -1
                print("-")
            else
                print("+")
            end
            if energy_preserving_conjugate == rootedtree(t)
                println("self")
            else
                println(energy_preserving_conjugate.level_sequence)
            end
            if !(energy_preserving_conjugate in treelist)
                push!(treelist,copy(energy_preserving_conjugate))
            end
        end
        println("")
    end
end
```
